```python
# This shows: Frequency of values in a dataset and also how data is distibuted
# Kernel Density Estimate (KDE) curve on top of your histogram bars.
#  
import matplotlib.pyplot as plt
import seaborn as sns
data = [10,33.12,7,9,12,15,20,25,30,35]
sns.histplot(data, bins=6, kde=True)
plt.title("Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# This creates a smooth distibution curve instead of bars
sns .kdeplot(data,fill = True)
plt.title("KDE Plot")
plt.show()
```

```python
median = np.median(data)
print("Median:", median)
quartiles = np.percentile(data, [25, 50, 75])
print("Quartiles:", quartiles)
outliers = [x for x in data if x < quartiles[0] or x > quartiles[2]]
print("Outliers:", outliers)
sns.boxplot(data)
plt.title("Box Plot")
```

```python
# Regression Visualization in seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = pd.DataFrame({"hours_studied": [1, 2, 3, 4, 5], "exam_score": [50, 60, 70, 80, 90]})
sns.regplot(x="hours_studied", y="exam_score", data=data)
plt.title("Regression Plot")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data = {
    "hours":[1,2,3,4,5],
    "score":[50,55,65,70,80]
}

df = pd.DataFrame(data)
```

```python
X = df[['hours']]
y = df['score']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[6]])
print(prediction)
```

```python
# Visualize the regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel("Study Hours")
plt.ylabel("Score")
plt.show()
```

```python
# pie plot
import matplotlib.pyplot as plt

# Sample data for the pie plot
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]  # Proportions for each category
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # Explode 1st slice (Category A)

# Create the pie chart
plt.figure(figsize=(10, 5))
plt.pie(sizes, labels=labels,explode=explode, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Sample Pie Plot')
plt.show()
```
